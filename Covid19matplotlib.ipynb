**Covid19 basic visualisation**


```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcdefaults()
#plt.style.use('fivethirtyeight')


data = pd.read_csv('selectedsea.csv')

daycount = data['Day']

MY = data['Malaysia']
ID = data['Indonesia']
TH = data['Thailand']
SG = data['Singapore']
PH = data['Philippines']
BR = data['Brunei']

colors = ['#1f77b4','#aec7e8','#ff7f0e','#ffbb78','#2ca02c','#98df8a']

plt.plot(daycount, MY, color=colors[0], label='Malaysia')
plt.plot(daycount, ID, color=colors[1], label='Indonesia')
plt.plot(daycount, TH, color=colors[2], label='Thailand')
plt.plot(daycount, SG, color=colors[3], label='Singapore')
plt.plot(daycount, PH, color=colors[4], label='Phillapines')
plt.plot(daycount, BR, color=colors[5], label='Brunei')

plt.legend()
plt.title('Coronavirus Infections Selected SEA countries')
plt.ylabel('Number of Infections')
plt.xlabel('Number of Days')

plt.savefig("plot.png")

plt.show()
```

Done for first visualisation.
